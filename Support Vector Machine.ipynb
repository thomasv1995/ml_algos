## Soft-margin

Goal is to minimize: $$\lambda \lVert \mathbf {w} \rVert ^{2}+\left[{\frac {1}{n}}\sum _{i=1}^{n}\max \left(0,1-y_{i}(\mathbf {w} ^{\mathsf {T}}\mathbf {x} _{i}-b)\right)\right]$$

where $\lambda > 0$ determines the trade-off between increasing the margin size and ensuring that the ${\displaystyle \mathbf {x} _{i}}$lie on the correct side of the margin
